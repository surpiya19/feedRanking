Understanding how user engagement evolves over time is crucial for evaluating the sustained impact of a new feature or algorithm. 
By tracking daily engagement rates across control and variant groups, we can observe not only immediate effects but also trends that indicate long-term behavior change.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

We analyzed engagement data collected over 30 days from a synthetic TikTok-like platform. Engagement was defined as any interaction such as liking, sharing, or commenting on a video.  
We segmented users into control and variant groups to simulate an A/B test, then calculated the daily engagement rate for each group.

```python
# Load and inspect data
df = pd.read_csv('../data/tiktok_synthetic_data.csv')
print(df.head())
```

```python
# Convert 'date' column to datetime if not already
df['date'] = pd.to_datetime(df['date'])
```

```python
# Aggregate daily engagement rate by group
daily_engagement = df.groupby(['date', 'group'])['engaged'].mean().reset_index()
```

The line chart clearly shows these trends, with the variant group’s engagement line sitting above the control group’s throughout the timeline, and the gap widening slightly as days progress.

```python
plt.figure(figsize=(12,6))
sns.lineplot(data=daily_engagement, x='date', y='engaged', hue='group', marker='o')
plt.title('Daily Engagement Rate Over Time by Group')
plt.ylabel('Engagement Rate')
plt.xlabel('Date')
plt.xticks(rotation=45)
plt.legend(title='Group')
plt.tight_layout()
plt.show()
```

The plot shows that the variant group consistently outperforms the control group in daily engagement rate across the 30-day period. While there is some daily fluctuation, the variant’s engagement rate generally stays above that of the control, indicating a positive effect from the new feature or treatment.

The gap between variant and control engagement rates appears relatively stable with occasional peaks, suggesting that the uplift is sustained over time rather than being a one-time spike.

This sustained higher engagement in the variant group suggests the tested change is effective in driving more user interactions, supporting a potential full rollout or further investment.
